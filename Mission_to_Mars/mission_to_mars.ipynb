```python
#Dependencies
from bs4 import BeautifulSoup
import pandas as pd
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

### NASA Mars News

```python
# Scrape the Mars News Site and collect the latest News Title and Paragraph Text. 
# Assign the text to variables that you can reference later.
# Example:
# news_title = "NASA's Next Mars Mission to Investigate Interior of Red Planet"
# news_p = "Preparation of NASA's next spacecraft to Mars, InSight, has ramped up this summer, 
# on course for launch next May from Vandenberg Air Force Base in central California -- 
# the first interplanetary launch in history from America's West Coast."
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of the page to be scraped
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('div', class_='list_text')
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        news_title = result.find_all('div', class_='content_title')[0].text
        # Identify and return price of listing
        news_p = result.find_all('div', class_='article_teaser_body')[0].text
        

        # Print results only if title and paragraph are available
        if (news_title and news_p):
            print('-------------')
            print(news_title)
            print(news_p)
    except AttributeError as e:
        print(e)
```
